```python
import pandas as pd
import numpy
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole
IPythonConsole.ipython_useSVG = True
from mordred import Calculator, descriptors
from PyBioMed.PyProtein import Autocorrelation, Autocorrelation
```

```python
df = pd.read_csv('./data/original_dataset/unassemble.csv')
df1 = pd.read_csv('./data/original_dataset/assemble.csv')
```

```python
pro=[]
for i in range(len(df)):
    output=Autocorrelation.CalculateMoranAutoTotal(df.loc[i,'sequence'])
    pro.append(output)
df2_0 = pd.DataFrame(pro)
df2_0.insert(0, 'sequence', df['sequence'])
#df2_0.to_csv('./data/preprocess_dataset/dataset_2class_MoranAutoTotal_unassemble.csv')
```

```python
pro1=[]
for i in range(len(df1)):
    output=Autocorrelation.CalculateMoranAutoTotal(df1.loc[i,'sequence'])
    pro1.append(output)
df2_1 = pd.DataFrame(pro1)
df2_1.insert(0, 'sequence', df1['sequence'])
#df2_1.to_csv('./data/preprocess_dataset/dataset_2class_MoranAutoTotal_assemble.csv')
```

```python
pro=[]
for i in range(len(df)):
    output=Autocorrelation.CalculateGearyAutoTotal(df.loc[i,'sequence'])
    pro.append(output)
df3_0 = pd.DataFrame(pro)
df3_0.insert(0, 'sequence', df['sequence'])
#df3_0.to_csv('./data/preprocess_dataset/dataset_2class_GearyAutoTotal_unassemble.csv')
```

```python
pro1=[]
for i in range(len(df1)):
    output=Autocorrelation.CalculateGearyAutoTotal(df1.loc[i,'sequence'])
    pro1.append(output)
df3_1 = pd.DataFrame(pro1)
df3_1.insert(0, 'sequence', df1['sequence'])
#df3_1.to_csv('./data/preprocess_dataset/dataset_2class_GearyAutoTotal_assemble.csv')
```

```python
pro=[]
for i in range(len(df)):
    output=Chem.MolToSmiles(Chem.MolFromFASTA(df.loc[i,'sequence']))
    pro.append(output)
calc = Calculator(descriptors, ignore_3D=False)
len(calc.descriptors)
mols = [Chem.MolFromSmiles(smi) for smi in pro]
df4_0 = calc.pandas(mols)
df4_0.insert(0, 'sequence', df['sequence'])
#df4_0.to_csv('./data/preprocess_dataset/dataset_2class_unassemble.csv')
```

```python
pro1=[]
for i in range(len(df1)):
    output=Chem.MolToSmiles(Chem.MolFromFASTA(df1.loc[i,'sequence']))
    pro1.append(output)
calc = Calculator(descriptors, ignore_3D=False)
len(calc.descriptors)
mols1 = [Chem.MolFromSmiles(smi) for smi in pro1]
df4_1 = calc.pandas(mols1)
df4_1.insert(0, 'sequence', df1['sequence'])
#df4_1.to_csv('./data/preprocess_dataset/dataset_2class_assemble.csv')
```

```python
df4_0['label']=0
df4_1['label']=1
df4_all=pd.concat([df4_0, df4_1],axis=0)
df5_all=df4_all.loc[:, (df4_all != 0).any(axis=0)]#data cleansing
df6_all = df5_all.select_dtypes(exclude=['object'])
df7_all=df6_all.sample(frac=1).reset_index(drop=True)
df7_all.to_csv('./data/preprocess_dataset/dataset_2class_1descriptor.csv')# mordred
```

```python
df_pio_0 = pd.merge(df2_0, df3_0, on='sequence')
df_pio_1 = pd.merge(df2_1, df3_1, on='sequence')
df_pio_0['label']=0
df_pio_1['label']=1
df4_all=pd.concat([df_pio_0, df_pio_1],axis=0)
df5_all=df4_all.loc[:, (df4_all != 0).any(axis=0)]#data cleansing
df6_all = df5_all.select_dtypes(exclude=['object'])
df7_all=df6_all.sample(frac=1).reset_index(drop=True)
df7_all.to_csv('./data/preprocess_dataset/dataset_2class_1descriptor_onlypio.csv')# piomed
```

```python
def merge_dataframes(df1, df2, df3, index_col):
    merged_df = pd.merge(df1, df2, on=index_col)
    merged_df = pd.merge(merged_df, df3, on=index_col)
    return merged_df

index_col = 'sequence'
df_0 = merge_dataframes(df2_0, df3_0, df4_0, index_col)
df_1 = merge_dataframes(df2_1, df3_1, df4_1, index_col)
```
