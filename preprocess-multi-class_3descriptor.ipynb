```python
import pandas as pd
import numpy
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole
IPythonConsole.ipython_useSVG = True
from mordred import Calculator, descriptors
from PyBioMed.PyProtein import Autocorrelation, Autocorrelation
```

```python
def process(fasta):
    peptide_smiles = Chem.MolToSmiles(Chem.MolFromFASTA(fasta))
    return peptide_smiles
```

```python
df = pd.read_csv('./data/original_dataset/data_all_3descriptor.csv')
#df['sequence']
pro=[]
for i in range(len(df)):
    print(i+2)
    print(df.loc[i,'sequence'])
    output=Autocorrelation.CalculateMoranAutoTotal(df.loc[i,'sequence'])
    pro.append(output)
df2_0 = pd.DataFrame(pro)
df2_0.insert(0, 'sequence', df['sequence'])

pro=[]
for i in range(len(df)):
    output=Autocorrelation.CalculateGearyAutoTotal(df.loc[i,'sequence'])
    pro.append(output)
df3_0 = pd.DataFrame(pro)
df3_0.insert(0, 'sequence', df['sequence'])

pro=[]
for i in range(len(df)):
    output=Chem.MolToSmiles(Chem.MolFromFASTA(df.loc[i,'sequence']))
    pro.append(output)
calc = Calculator(descriptors, ignore_3D=False)
len(calc.descriptors)
mols = [Chem.MolFromSmiles(smi) for smi in pro]
df4_0 = calc.pandas(mols)
df4_0.insert(0, 'sequence', df['sequence'])
```

```python
import numpy as np
amino_acids = 'ACDEFGHIKLMNPQRSTVWY'
sequences=df['sequence']
aa_to_index = {aa: i for i, aa in enumerate(amino_acids)}
num_aa = len(amino_acids)
max_length=41
encoded_sequences = []
for sequence in sequences:
    encoded_sequence = np.zeros((max_length, num_aa))
    for i, aa in enumerate(sequence):
        encoded_sequence[i, aa_to_index[aa]] = 1
    encoded_sequence_vector=encoded_sequence.flatten()
    encoded_sequences.append(encoded_sequence_vector)
df5_0 = pd.DataFrame(encoded_sequences)
df5_0.insert(0, 'sequence', df['sequence'])
```

```python
def merge_dataframes1(df1, df2, df3, df4, index_col):
    merged_df = pd.merge(df1, df2, on=index_col)
    merged_df = pd.merge(merged_df, df3, on=index_col)
    merged_df = pd.merge(merged_df, df4, on=index_col)
    return merged_df

index_col = 'sequence'
df_all_0 = merge_dataframes1(df2_0, df3_0, df4_0, df5_0, index_col)

df5_all_all=df_all_0.loc[:, (df_all_0 != 0).any(axis=0)]#data cleansing
df6_all_all = df5_all_all.select_dtypes(exclude=['object'])
#df = pd.read_csv('./data/original_dataset/data_all_3descriptor.csv')
df6_all_all['labels']=df['labels']
df7_all_all=df6_all_all.sample(frac=1).reset_index(drop=True)
df7_all_all.to_csv('./data/preprocess_dataset/dataset_multi-class_3descriptor.csv')
```
